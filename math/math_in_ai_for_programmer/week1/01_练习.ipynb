```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

x = np.linspace(-2, 2, 50)
# y = np.linspace(-2, 2, 50)
y = x.copy()

X, Y = np.meshgrid(x, y)
Z = np.exp(-(X**2+Y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap=matplotlib.cm.hot)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
