```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([3, 5])
b = np.array([2, 1])
c = a + b

V = np.array([a, b, c, b])
# print(V[:, 0])

plt.xlim(0, 10)
plt.ylim(0, 10)
plt.quiver([0, 0, 0, 3], [0, 0, 0, 5], V[:, 0], V[:, 1], color=['red', 'blue', 'black', 'blue'], angles='xy', scale_units='xy', scale=1)
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([3, 5])
b = np.array([2, 1])
c = a + b

print(tuple(a))

plt.xlim(0, 10)
plt.ylim(0, 10)
plt.annotate('', xy=tuple(a), xytext=(0, 0), arrowprops={'color': 'blue', 'width': 1})
plt.annotate('', xy=tuple(b), xytext=(0, 0), arrowprops={'color': 'green', 'width': 1})
plt.annotate('', xy=tuple(c), xytext=(0, 0), arrowprops={'color': 'black', 'width': 1})
plt.annotate('', xy=tuple(c), xytext=tuple(a), arrowprops={'color': 'green', 'width': 1})
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

origin = (0, 0)

fig, axs = plt.subplots(2, 2)
print(fig, axs[0][0])

a = np.array([3, 0])
b = np.array([1.414, 1.414])
c = a @ b
print('c1', c)
print('b len', np.linalg.norm(b, 2))
ax = axs[0][0]

ax.set_xlim(-2, 3)
ax.set_ylim(-2, 3)
ax.annotate('', xy=a, xytext=origin, arrowprops = {'color': 'blue', 'width': 1})
ax.annotate('', xy=b, xytext=origin, arrowprops = {'color': 'green', 'width': 1})

a = np.array([3, 0])
b = np.array([0, 2])
c = a @ b
print('c2', c)
print('b len', np.linalg.norm(b, 2))
ax = axs[0][1]

ax.set_xlim(-2, 3)
ax.set_ylim(-2, 3)
ax.annotate('', xy=a, xytext=origin, arrowprops = {'color': 'blue', 'width': 1})
ax.annotate('', xy=b, xytext=origin, arrowprops = {'color': 'green', 'width': 1})


a = np.array([3, 0])
b = np.array([-1.414, 1.414])
c = a @ b
print('c3', c)
print('b len', np.linalg.norm(b, 2))
ax = axs[1][0]

ax.set_xlim(-2, 3)
ax.set_ylim(-2, 3)
ax.annotate('', xy=a, xytext=origin, arrowprops = {'color': 'blue', 'width': 1})
ax.annotate('', xy=b, xytext=origin, arrowprops = {'color': 'green', 'width': 1})


a = np.array([3, 0])
b = np.array([2, 0])
c = a @ b
print('c4', c)
print('b len', np.linalg.norm(b, 2))
ax = axs[1][1]

ax.set_xlim(-2, 3)
ax.set_ylim(-2, 3)
ax.annotate('', xy=a, xytext=origin, arrowprops = {'color': 'blue', 'width': 1})
ax.annotate('', xy=b, xytext=origin, arrowprops = {'color': 'green', 'width': 1})


plt.show()
```
