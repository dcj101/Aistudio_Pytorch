```python
import numpy as np

AU = 1.495978707e8
Ms = 1.989e30
Rs = 500 * AU
Ps = 0.5

PlanetsInfo = [[0.3871, 3.30e23],
    [0.7233, 4.87e24],
    [1, 5.97e24],
    [1.5237, 6.42e23],
    [5.2026, 1.90e27],
    [9.5549, 5.68e26],
    [19.2184, 8.68e25],
    [30.1104, 1.02e26]]

def r2vol(r):
    return 3/4 * np.pi * r**3

def rh(alpha, mass):
    return alpha * (mass / (3 * Ms))**(1/3)

Vs = r2vol(500)

probs = []
for planet in PlanetsInfo:
    dist, mass = planet
    r = rh(dist, mass)
    vol = r2vol(r)
    
    prob = vol / Vs
    probs.append(prob)
    
P = Ps * np.sum(probs)
print(P)
```
