# 蒙提霍尔问题

```python
import random

def monty_hall(change_door=True, num_trials=10000):
    wins = 0
    for _ in range(num_trials):
        prize_door = random.randint(1, 3)
        chosen_door = random.randint(1, 3)
        
        if prize_door != 1 and chosen_door != 1:
            open_door = 1
        elif prize_door != 2 and chosen_door != 2:
            open_door = 2
        else:
            open_door = 3
        
        if change_door:
            final_choice = set([1, 2, 3]) - {chosen_door, open_door}
            final_choice = final_choice.pop()
        else:
            final_choice = chosen_door
        
        if final_choice == prize_door:
            wins += 1
    
    return wins

wins_without_changing = monty_hall(change_door=False, num_trials=10000)
print("不换门的中奖次数:", wins_without_changing)
print("不换门的中奖概率:", wins_without_changing / 10000)

wins_with_changing = monty_hall(change_door=True, num_trials=10000)
print("换门的中奖次数:", wins_with_changing)
print("换门的中奖概率:", wins_with_changing / 10000)
```
