```python
import numpy as np

t = np.sqrt(5)

v1 = np.array([2, t-1])
v1norm = np.linalg.norm(v1)

v1 = v1 / v1norm
print(v1)

v2 = np.array([2, -t-1])
v2norm = np.linalg.norm(v2)

v2 = v2 / v2norm
print(v2)

l1 = 4 + t
l2 = 4 - t

print(l1, l2)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 100, 20)
y = np.linspace(0, 100, 20)

X, Y = np.meshgrid(x, y)

A = np.array([4, -1, 0, 2]).reshape(2, 2)
D = (X * 4 - Y)
L = 2 * Y

norm = np.sqrt(D**2 + L**2)
Dn = D / norm
Ln = L / norm

vals, vects = np.linalg.eig(A)
print(vects[:, 1])

v1 = vects[:, 0] * 100
v2 = vects[:, 1] * 100

print(v1 @ v2)

fig, ax = plt.subplots()
ax.quiver(X, Y, Dn, Ln, units='xy', scale=0.2, angles='uv', color='blue')

ax.plot([0, v1[0]], [0, v1[1]])
ax.plot([0, v2[0]], [0, v2[1]])

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-50, 50, 20)
y = np.linspace(-50, 50, 20)

X, Y = np.meshgrid(x, y)

A = np.array([0, -1, 1, 0]).reshape(2, 2)
X2 = -Y
Y2 = X

norm = np.sqrt(X2**2 + Y2**2)
X2 = X2 / norm
Y2 = Y2 / norm

vals, vects = np.linalg.eig(A)
print(vals)
print(vects)

v1 = vects[:, 0] * 100
v2 = vects[:, 1] * 100

fig, ax = plt.subplots()
ax.quiver(X, Y, X2, Y2, units='xy', scale=0.2, angles='uv', color='blue')

ax.plot([0, v1[0]], [0, v1[1]])
ax.plot([0, v2[0]], [0, v2[1]])

ax.grid(True)
ax.set_aspect('equal')
plt.show()
```
