```python
import sympy as sp

x = sp.symbols('x')

f = 3 * x ** 2 + 2 * x + 1

f_int = sp.integrate(f, x)
display(f_int)
```

```python
import scipy

def f(x):
    return x ** 2

result, error = scipy.integrate.quad(f, 0, 1)

print(result, error)
```

```python
import sympy as sp

x = sp.symbols('x')

f = x ** 2

f_int = sp.integrate(f, (x, 0, 1))
display(f_int)
```

```python
import scipy
import numpy as np

def f(x):
    return np.sin(x)

result, error = scipy.integrate.quad(f, 0, scipy.constants.pi)

print(result, error)
```

```python
import scipy
import numpy as np

def f(x):
    return 1 / x

result, error = scipy.integrate.quad(f, 1, np.e)

print(result, error)
```

```python
import sympy as sp

x = sp.symbols('x')

f = x * sp.exp(x)

f_int = sp.integrate(f, x)
display(f_int)
```

```python
import sympy as sp

x = sp.symbols('x')

f = x * sp.ln(x)

f_int = sp.integrate(f, x)
display(sp.simplify(f_int))
```

```python
import sympy as sp

x = sp.symbols('x')

f = sp.atan(x)

f_int = sp.integrate(f, x)
display(sp.simplify(f_int))
```

```python
import sympy as sp

x, a, b = sp.symbols('x a b')

f = b * sp.sqrt(1 - x**2/a**2)
f_int = sp.integrate(f, x)
display(sp.simplify(f_int))
```
