# 数值积分 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad


def myfunc(x):
    with np.errstate(divide='ignore', invalid='ignore'):
        return np.where(x == 0, 1, np.sin(x) / x)


result, error = quad(myfunc, -10, 10, epsabs=0.01, epsrel=0.01)
print(f"result：{result}, err：{error}")

x = np.linspace(-10, 10, 400)

y = myfunc(x)

plt.plot(x, y)
plt.fill_between(x, y, where=(y >= 0), facecolor='green', alpha=0.2)
plt.fill_between(x, y, where=(y < 0), facecolor='red', alpha=0.2)

plt.show()
```

# sympy解析计算积分

```python
import sympy as sp

x = sp.symbols('x')

integral = sp.integrate(sp.sin(x)/x, x)
display(integral)

result1 = sp.N(integral.subs(x, -10))
result2 = sp.N(integral.subs(x, 10))
display(result2 - result1)
```

```python
import sympy as sp

x = sp.symbols('x')

integral = sp.integrate(sp.sin(x)/x, (x, -10, 10))
display(integral)

numerical_integral = sp.N(integral)
print(numerical_integral)
```
