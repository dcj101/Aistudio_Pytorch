通过逆矩阵求线性方程组的解

```python
import numpy as np

A = np.array([
    [2, 1, -1, 2],
    [-3, -1, 2, -4],
    [-2, 1, 2, -3],
    [1, 1, -2, 5],
])

b = np.array([8, -11, -3, 9])

# print(np.linalg.det(A))
Ainv = np.linalg.inv(A)
x = Ainv @ b

print(np.round(x, 2))
```

计算矩阵的秩：判断有多少个线性不相关的向量

```python
import numpy as np

A = np.array([
    [1, 2], 
    [-3, -6]
])

print(np.linalg.matrix_rank(A))
```

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([
    [1, 2], 
    [-3, -6]
])

plt.xlim(-3, 3)
plt.ylim(-3, 3)

plt.quiver([0, 0], [0, 0], A[:, 0], A[:, 1], color=['red', 'green'], angles='xy', scale_units='xy', scale=1)
plt.grid()
plt.show()

```

```python
import numpy as np

A = np.array([
    [1,2,3],
    [4,5,6]
])

B = np.array([
    [7,8],
    [9,10],
    [11,12]
])

print(np.tensordot(A, B, axes=0))
```
