```python
import numpy as np

# 定义矩阵A和向量b
A = np.array([[1, 6],
              [2, -1],
              [3, 1],
              [5, -3],
              [6, 1],
              [7, -2]])

b = np.array([1, 9, 5, 3, 4, 1])

# 使用NumPy的linalg.qr函数进行QR分解
Q, R = np.linalg.qr(A)
print(Q)
print(R)

# 计算QTb
Q_T_b = Q.T @ b

# 求解上三角系统RTRx = RTQTb
# 由于R是上三角矩阵，RT也是上三角矩阵，因此RTR仍然是上三角矩阵
# 可以使用NumPy的linalg.solve函数直接求解
x = np.linalg.solve(R.T @ R, R.T @ Q_T_b)

# 打印结果
print("最小二乘解x：")
print(x)

print(b - A @ x)
```

```python
import numpy as np

A = np.array([
    [0.6, 10],
	[0.29, 15],
	[0.7, 7],
	[0.55, 6],
	[0.59, 7],
	[0.67, 6],
	[0.4, 10],
	[0.89, 6],
])

print(np.linalg.matrix_rank(A))

b = np.array([6,8,3,5,5,5,7,5])

ATA = A.T @ A
ATb = A.T @ b
L = np.linalg.cholesky(ATA)
print(ATA.shape)
print(ATb.shape)
y = np.linalg.solve(L, ATb)

x = np.linalg.solve(L.T, y)
print(x)



```

```python
import numpy as np

A = np.array([
    [0.6, 10],
	[0.29, 15],
	[0.7, 7],
	[0.55, 6],
	[0.59, 7],
	[0.67, 6],
	[0.4, 10],
	[0.89, 6],
])

b = np.array([6,8,3,5,5,5,7,5])

Q, R = np.linalg.qr(A)
c = Q.T @ b

x = np.linalg.solve(R, c)

print(c)
print(R @ x)
```
