```python
from scipy.stats import norm
import numpy as np

mu = 55
sigma = 5

good_prob = 1 - norm.cdf(58, mu, sigma)
good_num = np.int32(np.round(1000 * good_prob))
print(f'优选鸡蛋的数量： {good_num}')

unqualified_prob = norm.cdf(52, mu, sigma)
unqualified_num = np.int32(unqualified_prob * 1000)
print(f'不合格鸡蛋的数量： {unqualified_num}')

target_prop = 350 / 1000
z = norm.ppf(1 - target_prop)

w = np.int32(np.round(z * sigma + mu))
print(w)

w2 = norm.isf(target_prop, mu, sigma)
print(w2)
```

```python
from scipy.stats import expon
import numpy as np

mu = 5
lmd = 1 / mu

prob_less_3 = expon.cdf(3, scale=1/lmd)
print(f"处理时间小于3分钟的概率：{prob_less_3}")

prob_more_10 = 1 - expon.cdf(10, scale=1/lmd)
print(f"处理时间大于10分钟的概率：{prob_more_10}")

lmd1 = 0.7667
print(expon.cdf(3, scale=1/lmd1))
```

```python
from scipy.stats import poisson

lambda_rate = 60/5

prob_20 = poisson.pmf(20, mu=lambda_rate)
print(f'一个小时20订单的概率：{prob_20}')

prob_6 = poisson.pmf(6, mu=lambda_rate)
print(f'一个小时6订单的概率：{prob_6}')

prob_at_least_15 = 1 - poisson.cdf(14, mu=lambda_rate)
print(f'一个小时至少15订单的概率：{prob_at_least_15}')
```

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

np.random.seed(0)
data = stats.skewnorm.rvs(a=10, loc=0, scale=100, size=10000)
data = np.round(data)

mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data).mode
print(mode)
print(np.where(data == 27)[0].size)
print(mean, median, mode)




plt.hist(data, bins=50, edgecolor='black')

plt.axvline(mean, label='mean', color='green')
plt.axvline(median, label='median', color='red')
plt.axvline(mode, label='mode', color='purple')

plt.legend()
plt.show()
```
