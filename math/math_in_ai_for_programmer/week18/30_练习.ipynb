```python
import numpy as np

from scipy.optimize import minimize

true_mu = 5
true_sigma = 1.5
sample_size = 10000

np.random.seed(20)
data = np.random.normal(true_mu, true_sigma, sample_size)
n = len(data)

def log_like(params):
    mu, sigma = params
    
    rt = n / 2 * np.log(2 * np.pi * sigma**2) + 1 / (2 * sigma**2) * np.sum((data - mu)**2)
    return rt

result = minimize(log_like, [0, 1])

est_mu, est_sigma = result.x
print(f'est mu {est_mu}, est sigma {est_sigma}')
print(f'data mu {np.mean(data)}, data sigma {np.std(data)}')
```
