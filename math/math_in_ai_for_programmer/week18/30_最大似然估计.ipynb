# 最大似然估计逻辑回归

```python
import numpy as np

def sigma(z):
    return 1 / (1 + np.exp(-z))

def likelihood(w, b, X, y):
    n = X.shape[0]
    
    z = np.dot(X, w) + b
    p = sigma(z)
    return np.sum(y * np.log(p) + (1 - y) * np.log(1 - p)) / n

def compute_gradient(w, b, X, y):
    z = np.dot(X, w) + b
    p = sigma(z)

    n = X.shape[0]
    dw = np.dot(X.T, (p - y))/n
    db = np.sum(p - y)/n
    return (dw, db)

def gradient_descent(X, y, init_w, init_b, learning_rate, max_iter):
    w = init_w
    b = init_b

    for i in range(max_iter):
        (dw, db) = compute_gradient(w, b, X, y)

        w = w - learning_rate * dw
        b = b - learning_rate * db

        if(i % 1000 == 0):
            print(f'iter {i}, 似然：{likelihood(w, b, X, y)}')

    return (w, b)


X = np.array([[0.5, 1], [1.5, 2], [3, 4], [4, 5], [5, 6]])
y = np.array([0, 0, 1, 1, 1])

init_w = np.zeros(X.shape[1])
init_b = 0

learning_rate = 0.05
max_iter = 10000

w, b = gradient_descent(X, y, init_w, init_b, learning_rate, max_iter)
print(f'最后似然：{likelihood(w, b, X, y)}')
print("\n参数w：", w, ' b:', b)

z = X.dot(w) + b
p = sigma(z)
print(f'\n概率：{p}')
print(f'估计标签：{np.round(p).astype(int)}\n真实标签：{y}')
```
