```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

x = np.linspace(0, 3, 100)
y = f(x)

x1 = 1
x2 = 1.01
y1 = f(x1)
y2 = f(x2)

slop = (y2 - y1) / (x2 - x1)
y_plus = y1 + slop * (x - x1)


plt.figure()

plt.plot(x, y, label = 'y=x^2')
# plt.plot([x1, x2], [y1, y2])
plt.plot(x, y_plus, label = 'line through 2 points')
plt.plot(x, y1 + 2 * (x - x1), label = 'y=2x')

plt.legend()
plt.axis('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

def df(x):
    return 2 * x

x = np.linspace(0, 3, 100)
x2 = x + 0.01
y = f(x)
y2 = f(x2)

slop = (y2 - y) / (x2 - x)

fig, ax = plt.subplots()
ax.plot(x, slop, label='slop')
ax.plot(x, df(x), label='y=2x')

plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

def area_sum_by_rectangles(start_x, end_x):
    n_per_unit = 50
    n = n_per_unit * (end_x - start_x)
    
    width = 1 / n_per_unit
    
    x_values = np.linspace(start_x, end_x, n+1)
    
    
    area_sum = 0
    for i in range(n):
        area_sum += width * f(x_values[i])
    
    return area_sum


n_plots = 10
areas = []
for end_x in range(1, n_plots + 1):
    area = area_sum_by_rectangles(0, end_x)
    areas.append(area)
    
fig, ax = plt.subplots()
ax.plot(np.arange(1, n_plots+1), areas)

x = np.arange(1, n_plots+1)
ax.plot(x, (x**3)/3)


# plt.axis('equal')
plt.show()
    
```
