```python
import sympy as sp

x = sp.symbols('x')
f = (x**2-1)/(x-1)

limit_value = sp.limit(f, x, 1)

print(limit_value)
```

```python
import sympy as sp

x = sp.symbols('x')

f = sp.Piecewise((x**2, x<=1), (2*x+1, x > 1))

left_limit = sp.limit(f, x, 1, dir='-')
right_limit = sp.limit(f, x, 1, dir='+')

print(left_limit, right_limit)

from sympy.plotting import plot 
plot(f, xlim=(-1, 2), ylim=(0, 5))
```

```python
import sympy as sp

x = sp.symbols('x')
f = sp.sin(x) /x

limit_value = sp.limit(f, x, 0)

print(limit_value)

from sympy.plotting import plot 
plot(f)
```

```python
import sympy as sp

x = sp.symbols('x')
f = (sp.exp(x) - 1) / x

limit_value = sp.limit(f, x, 0)

print(limit_value)

from sympy.plotting import plot 
plot(f, xlim=(-1, 1), ylim=(0, 5))
```

```python
import sympy as sp

x = sp.symbols('x')
f = sp.ln(1 + sp.sin(x))

limit_value = sp.limit(f, x, 0)

print(limit_value)

from sympy.plotting import plot 
plot(f, ylim=(-2, 2))
```
