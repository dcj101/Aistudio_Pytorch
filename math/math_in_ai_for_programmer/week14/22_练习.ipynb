```python
import torch

x = torch.tensor(2.0, requires_grad=True)

def sigmoid(x):
    return 1 / (1 + torch.exp(-x))

y = sigmoid(x)
print(y)

y.backward()

x.grad
```

```python
import torch

x = torch.tensor([2.0], requires_grad=True)

y = x**2
z = 3 * y + 1

z.backward()
print(x.grad)

y.grad_fn
```

```python
def print_comp_grah(tensor, indent=0):
    print(' ' * indent + str(tensor))
    if hasattr(tensor, 'grad_fn') and tensor.grad_fn is not None:
        print(' ' * (indent + 2) + str(tensor.grad_fn))
        for funcs in tensor.grad_fn.next_functions:
            if funcs is None: continue
            print_comp_grah(funcs, indent + 4)

x = torch.tensor([2.0], requires_grad=True)

y = x**2
z = 3 * y + 1
z.backward()
print_comp_grah(z)
```
