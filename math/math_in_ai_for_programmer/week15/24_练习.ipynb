```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(10)

def generate_spiral_data(N=300, K=3):
    X = np.zeros((N * K, 2))
    y = np.zeros(N * K, dtype='uint8')
    for j in range(K):
        ix = range(N * j, N * (j + 1))
        r = np.linspace(0.0, 1, N)
        t = np.linspace(j * 4, (j + 1) * 4, N) + np.random.randn(N) * 0.2
        X[ix] = np.c_[r * np.sin(t), r * np.cos(t)]
        y[ix] = j
    return X, y

X, y = generate_spiral_data()

fig, ax = plt.subplots()
# for i in range(3):
#     mask = i == y
#     ax.scatter(X[mask, 0], X[mask, 1])
ax.scatter(X[:, 0], X[:, 1], c=y)

plt.show()
```

```python
import sklearn



X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, test_size=0.2, random_state=10)

for k in ['linear', 'poly', 'rbf']:
    if k == 'poly':
        svm = sklearn.svm.SVC(kernel=k, C=0.1, degree=3)
    else:
        svm = sklearn.svm.SVC(kernel=k, C=0.1)

    svm.fit(X_train, y_train)
    y_pred = svm.predict(X_test)
    acc = sklearn.metrics.accuracy_score(y_test, y_pred)

    print(f'kernel {k}, acc {acc}')
```
