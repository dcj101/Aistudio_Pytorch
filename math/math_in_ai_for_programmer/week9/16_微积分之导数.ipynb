# 常见函数的导数

```python
from sympy import *

x, n, a = symbols('x, n, a')

funs = [x**n, E**x, a**x, ln(x), log(x, a), 
        sin(x), cos(x), tan(x), cot(x), sec(x), csc(x),
        asin(x),acos(x),atan(x),acsc(x),asec(x),acot(x)
    ]

ps = None
for f in funs:
    dr = Derivative(f, x)
    display(Eq(dr ,simplify(diff(f, x))))
```

# 导数规则

```python
from sympy import *


x, c = symbols('x, c')
u, v = symbols('u v', cls=Function)

funs = [u(x) - v(x), c * u(x), u(x)*v(x), u(x)/v(x)]

for f in funs:
    dr = Derivative(f, x)
    display(Eq(dr ,simplify(diff(f, x))))
```

# 洛必达法则

```python
from sympy import *

x = symbols('x')
plot(sin(x)/x, (x, -20, 20))

d1 = diff(sin(x), x)
print(d1)
l1 = limit(d1, x, 0)
print(l1)

d2 = diff(x, x)
print(d2)

print(limit(sin(x)/x, x, 0))
```

## 链式法则在神经网络中的作用

### 激活函数

```python
from sympy import *

x, w, b = symbols('x, w, b')

f = 1 / (1 + E**(-x))
display(f)
plot(f)

d = diff(f, x)
display(d)
print(d.equals(f * (1 - f)))

z = w*x + b
display(diff(z, w))
display(diff(z, b))
```
