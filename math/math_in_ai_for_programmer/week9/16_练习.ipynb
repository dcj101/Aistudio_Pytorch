```python
import sympy as sp

x = sp.symbols('x')

f = (sp.exp(x) - x - 1) / x**2

print(sp.limit(f, x, 0))

f1 = sp.exp(x) - x - 1
f2 = x**2

f1_limit = sp.limit(f1, x, 0)
f2_limit = sp.limit(f2, x, 0)

while f1_limit == f2_limit and (f1_limit == 0 or f1_limit == sp.oo):
    f1 = sp.diff(f1, x)
    f2 = sp.diff(f2, x)

    f1_limit = sp.limit(f1, x, 0)
    f2_limit = sp.limit(f2, x, 0)

print(f1_limit/f2_limit)
```
