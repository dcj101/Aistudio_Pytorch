```python
import sympy as sp

x, y = sp.symbols('x y')
z = sp.ln(sp.sin(x - 2 * y))

display(sp.simplify(sp.diff(z, x)))
display(sp.simplify(sp.diff(z, y)))
```

```python
import numpy as np

epochs = 500

x = 2
y = 3
target = 8

learning_rate = 0.002

np.random.seed(42)
w1 = np.random.randn()
w2 = np.random.randn()
w3 = np.random.randn()
b1 = np.random.randn()

loss_hist = []
for epoch in range(epochs):
    z1 = x * w1 + b1
    z2 = z1 ** 2 + w2 * y
    z3 = np.sin(z2) + w3 * z1
    
    loss = (z3 - target)**2
    loss_hist.append(z3 - target)
    
    dz3_dw3 = z1
    dloss_dz3 = 2 * (z3 - target)
    dloss_dw3 = dloss_dz3 * dz3_dw3
    
    dz3_dz2 = np.cos(z2)
    dz2_dw2 = y
    dloss_dw2 = dloss_dz3 * dz3_dz2 * dz2_dw2
    
    dz1_dw1 = x
    
    dz3_dz1 = w3
    dz2_dz1 = 2 * z1
    dloss_dz1 = dloss_dz3 * dz3_dz1 + dloss_dz3 * dz3_dz2 * dz2_dz1
    
    dloss_dw1 = dloss_dz1 * dz1_dw1
    dloss_db1 = dloss_dz1
    
    b1 -= learning_rate * dloss_db1
    w1 -= learning_rate * dloss_dw1
    w2 -= learning_rate * dloss_dw2
    w3 -= learning_rate * dloss_dw3
    
    if epoch % 50:
        print(f"epoch {epoch} loss: {loss}")
    
print(b1, w1, w2, w3)
print('z3=', z3)
```

```python
import matplotlib.pyplot as plt

temp = loss_hist[:50]

fig, ax = plt.subplots()
ax.plot(np.arange(len(temp)), temp)
plt.show()
```
